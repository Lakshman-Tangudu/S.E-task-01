```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, coeffs):
    a = float(coeffs[0])
    b = float(coeffs[1])
    c = float(coeffs[2])
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    with open("file1.txt", 'r') as f:
        lines = f.readlines()

    for idx, line in enumerate(lines):
        coeffs = line.strip().split(",")
        temp = quadratic_model(time_values, coeffs)
        plt.plot(time_values, temp, label=f'Set {idx + 1} {coeffs}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Input from file)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)
    a = float(input("Enter a value : "))
    b = float(input("Enter b value : "))
    c = float(input("Enter c value : "))
    listval = [[40, -3, 10], [a, b, c]]

    for i in listval:
        temp = quadratic_model(time_values, i[0], i[1], i[2])
        if i == [40, -3, 10]:
            plt.plot(time_values, temp, label=f'Hardcoded {i}')
        else:
            plt.plot(time_values, temp, label=f'User Input {i}')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Combination)')
    plt.show()

if __name__ == "__main__":
    main()
```
